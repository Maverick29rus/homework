**Шаг 4: Оценка множителей трудоемкости (Cost Drivers или Effort Multiplier)**

**Шаг 5: Расчеты с использованием COCOMO II**

Используем следующие формулы для расчетов:

1. **Формула оценки трудоемкости проекта в чел.$\times$мес.:**

  Посчитаем трудозатраты по упрощенной формуле:

  $PM = A \times (Size)^B \times EAF = 2.4 \times (0.3)^{1.05} \times (1.26 \times 1.22 \times 0.49 \times 0.95 \times 0.87 \times 1 \times 1) = 0.422{(чел.}\times \text{мес.) }$

2. **Фактор коррекции трудозатрат (E):**

  В упрощенной формуле не расчитывается

3. **EAF (Effort Adjustment Factor) - произведение выбранных множителей трудоемкости:**

  $EAF = ∏_{j=1}^nEM_j =  1.26 \times 1.22 \times 0.49 \times 0.95 \times 0.87 \times 1 \times 1 = 0.623$

  где $EM_j$ - множители трудоемкости (Effort Multiplier).
  n = 7 - для предварительной оценке

4. **Расчет времени разработки (TM, Time at Month):**

  Упрощенный вариант расчета времени разработки:
$TM = C \times PM^{D} = TM = 2.5 \times (0.422)^{0.38} = 1.8 \text{ мес. }$

5. **Расчет стоимости разработки (Development Cost)**:

  $DevelopmentCost = PM \times StaffSalary = 0.422 \times 100000 = 42200$,

6. $A, B, C, D$ - коэффициенты, зависящие от режима разработки.

  $B=1.05$, $C=2.5$, $D=0.38$.

  $A=2.4$ для предварительной оценки
